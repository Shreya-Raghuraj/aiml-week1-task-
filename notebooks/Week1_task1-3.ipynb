# Data cleaning process



```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df['gender'] = df['gender'].str.lower()
```

```python
df.rename(columns={
    'math score': 'math_score',
    'reading score': 'reading_score',
    'writing score': 'writing_score'
}, inplace=True)
```

```python
df.to_csv("cleaned_student_data.csv", index=False)
```

## Cleaning Summary

- No major missing values were found in the dataset.
- No duplicate rows were detected.
- Gender labels were standardized for consistency.
- Column names were cleaned for easier analysis.
- The cleaned dataset was saved separately.

# Data Visualization

```python
plt.figure(figsize=(8,5))

sns.histplot(df['math_score'], bins=20)

plt.title("Distribution of Math Scores")
plt.xlabel("Math Score")
plt.ylabel("Frequency")

plt.show()
```

## Visualization Insight

- Most students scored between 50 and 80 in mathematics.
- The distribution appears moderately spread.
- Very low and very high scores are less frequent.
